```python
# vamos a determinadar si la variable 'place_name' tiene un dato 'confiable' o 'completo', 
# para eso separamos la variable 'place_with_parent_names' en partes ... por el simbolo | (pipe) 
    
# nuevo df con el indice y la variable
datax = data[["place_with_parent_names"]]

datay = datax.join(datax["place_with_parent_names"].str.split('|', expand=True).rename(columns={1:'Pais', 
                                                                                                2:'Zona', 
                                                                                                3:'Partido_barrio', 
                                                                                                4:'Localidad', 
                                                                                                5:'Obs_localidad', 
                                                                                                6:'Descarte'}))
print(datay)
dataz = data.join(datay[["Pais","Zona","Partido_barrio","Localidad","Obs_localidad"]])
dataz
```

```python
# dependiendo de la zona que estamos analizando entendemos que la cantidad de info que necesitamos

valores_zona = ['Bs.As. G.B.A. Zona Norte','Bs.As. G.B.A. Zona Oeste','Bs.As. G.B.A. Zona Sur']
mask_Localidad = ([x in valores_zona for x in dataz.Zona]) & ((dataz.Localidad.isnull()) | (dataz.Localidad == ''))
dataz.loc[mask_Localidad, "calidad_dato"] = True
# caso especial de MDQ
valores_zona_mdq = ['Buenos Aires Costa Atlántica']
mask_Localidad_mdq = ([x in valores_zona_mdq for x in dataz.Zona]) & (dataz.Partido_barrio== 'Mar del Plata') & ((dataz.Localidad.isnull()) | (dataz.Localidad == ''))
dataz.loc[mask_Localidad, "calidad_dato"] = True
#resto_zonas
valores_zona_resto = ['Buenos Aires Costa Atlántica','Buenos Aires Interior','Capital Federal','Resto de provincias']
mask_Partido_barrio = ([x in valores_zona_resto for x in dataz.Zona]) & ((dataz.Partido_barrio.isnull()) | (dataz.Partido_barrio == ''))
mask_calidad_dato_false = dataz.calidad_dato == False
dataz.loc[mask_Partido_barrio & mask_calidad_dato_false, "calidad_dato"] = True

print(dataz.calidad_dato.value_counts())
mask_faltante =  (dataz.calidad_dato == True)
data_faltante = dataz.loc[mask_faltante]
#print (data_faltante)

groupby_df_faltante = data_faltante.groupby(['Pais',  'Zona', 'Partido_barrio'])["Id_caso"].count().sort_values(ascending=False)
print (groupby_df_faltante)
print (data_faltante.title.value_counts())

#ejemplo duplicado
mascara_ej = data_faltante.title =='Departamento en P.Chacabuco'

print(data_faltante.loc[mascara_ej])

```
